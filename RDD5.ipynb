```python
import pyspark
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```python
rdd1 = sc.textFile("/home/iceberg/notebooks/Data/StudentData.csv")
```

```python
rdd1.collect()
```

```python
rdd2 = rdd1.flatMap(lambda x: x.split(","))
rdd2.collect()
```

```python
rdd2.count()
```

```python
#wordcount Example
```

```python
#if you want to implemnt Groupbykey or Reducebykey
#1. Convert your data into k,v pairs
rdd3 = rdd2.map(lambda x: (x,1))
rdd3.collect()
```

```python
rdd3.groupByKey().mapValues(list).collect()
```

```python
def f(x): return len(x)
rdd3.groupByKey().mapValues(f).collect()
```
